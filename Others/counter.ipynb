```python
import os
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Specify the parent folder path
parent_folder_path = "/content/drive/MyDrive/Master's_Thesis/Dataset"  # Replace with the actual parent folder path

# Function to count image files in a folder
def count_images_in_folder(folder_path):
    # List all files in the folder and filter only images
    image_extensions = ('.png', '.jpg', '.jpeg', '.gif', '.bmp', '.tiff')
    files = os.listdir(folder_path)
    images = [file for file in files if file.lower().endswith(image_extensions)]
    return len(images)

# List to store folders with less than 1900 images
folders_less_than_1900 = []

# Loop through the subfolders named 1 to 50
for i in range(1, 51):
    subfolder_path = os.path.join(parent_folder_path, str(i))

    if os.path.exists(subfolder_path) and os.path.isdir(subfolder_path):
        # Count the number of images in the current subfolder
        image_count = count_images_in_folder(subfolder_path)

        # Print the folder and image count (optional)
        print(f'Folder {i} contains {image_count} images.')

        # If the folder contains fewer than 1900 images, store it in the list
        if image_count < 1901:
            folders_less_than_1900.append((str(i), image_count))

# Output all folders with less than 1900 images
if folders_less_than_1900:
    print('\nFolders with fewer than 1900 images:')
    for folder_name, image_count in folders_less_than_1900:
        print(f'Folder "{folder_name}" contains {image_count} images.')
else:
    print('No folder contains fewer than 1900 images.')
```
